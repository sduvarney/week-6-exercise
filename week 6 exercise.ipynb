```python
# add load_sales_data function to load sales data from a csv file
# add a docstring to the function to explain what it does and its parameters
import csv
def load_sales_data(file_path):
    """
    Load sales data from a CSV file.

    Parameters:
        file_path (str): The path to the CSV file containing sales data.

    Returns:
        list: A list of dictionaries representing each row in the CSV file.
    """
    sales_data = []
    with open(file_path, mode='r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            sales_data.append(row)
    return sales_data
```
